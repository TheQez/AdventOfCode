```python
f = open("input.txt")

zeros = 0
axis = 50

for line in f:
    is_right = line[0] == 'R';
    move = int(line[1:])
    if ( is_right ):
        axis += move
    else:
        axis -= move
    if ( axis % 100 == 0 ):
        zeros += 1

print(zeros)
f.close()
```

```python
f = open("input.txt")

zeros = 0
axis = 50
was_top = False
was_bottom = False

for line in f:
    is_right = line[0] == 'R';
    move = int(line[1:])
    last_axis = axis
    if ( is_right ):
        axis += move
    else:
        axis -= move

    zeros += abs( (axis // 100) - (last_axis // 100))

    if ( last_axis % 100 == 0 and
         not is_right ):
        zeros -= 1
    if ( axis % 100 == 0 and
         not is_right ):
        zeros += 1

print(zeros)
f.close()
```
