```python
f = open("input.txt")

input = []
for line in f:
    input.append([int(x) for x in line.rstrip('\n')])

f.close()
```

```python
def joltage( bank, num_digits ):
    head = 0
    sum = 0
    for i in range(num_digits-1, -1, -1):
        idx = max(range(head, len(bank)-i), key=bank.__getitem__)
        sum += (10**i)*bank[idx]
        head = idx+1

    return sum

joltage([3, 1, 2, 1, 7, 1, 4, 6], 2)
```

```python
total = 0
import time

for bank in input:
    total += joltage( bank, 12 )

print(total)
```
