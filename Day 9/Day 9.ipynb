```python
input = []

f = open("input.txt")

for line in f:
    input.append([int(num) for num in line.rstrip().split(',')])

f.close()

print(input)
```

```python
def area( x, y ):
    return (abs(x[0] - y[0])+1)*(abs(x[1] - y[1])+1)

area([1, 2], [5, 0])
```

```python
best_area = 0

for i in range (0, len(input)):
    for j in range(0, len(input)):
        if area( input[i], input[j] ) > best_area:
            best_area = area( input[i], input[j] )
            #print(best_area)

print(best_area)
```

```python
def itvl_x_itvl( itvl1, itvl2 ):
    #print(itvl1)
    #print(itvl2)
    return ( (itvl2[0] <= itvl1[0] and itvl1[0] <= itvl2[1]) or
             (itvl1[0] <= itvl2[0] and itvl2[0] <= itvl1[1]) )

def rect_x_line( rect_x, rect_y, line_x, line_y ):
    if itvl_x_itvl( rect_x, line_x ) and itvl_x_itvl( rect_y, line_y ): #Some clash
        rect_a = []
        rect_b = []
        line_a = []
        line_const = 0
        if line_x[0] == line_x[1]: #Line along x
            rect_a = rect_y
            line_a = line_y
            rect_b = rect_x
            line_const = line_x[0]
        else:
            rect_a = rect_x
            line_a = line_x
            rect_b = rect_y
            line_const = line_y[0]

        #print(rect_a)
        #print(rect_b)
        #print(line_a)
        #print(line_const)

        if line_const == rect_b[0] or line_const == rect_b[1]:
            return False #Skim

        if line_a[1] == rect_a[0] or line_a[0] == rect_a[1]:
            return False #Line ends on box

        return True
    return False #No intersection

rect_x_line([3, 5], [7, 9], [3, 3], [1, 10])
```

```python
def rect_is_admissible( coord1, coord2, input ):
    #print("Check rect", coord1, coord2)
    admissible = True

    low_x = min(coord1[0], coord2[0])
    high_x = max(coord1[0], coord2[0])
    low_y = min(coord1[1], coord2[1])
    high_y = max(coord1[1], coord2[1])

    # Check no corners inside
    for i in range(0, len(input)):
        #print("Interior?:", input[i][0], input[i][1])
        if ( low_x < input[i][0] and input[i][0] < high_x and
             low_y < input[i][1] and input[i][1] < high_y ):
            #print("Not admissible")
            admissible = False
            break

    # Check no line through the rectangle
    rect_x = [low_x, high_x]
    rect_y = [low_y, high_y]
    if admissible:
        for i in range(0, len(input)):
            line_x = [min(input[i][0], input[(i+1)%len(input)][0]),
                      max(input[i][0], input[(i+1)%len(input)][0]),]
            line_y = [min(input[i][1], input[(i+1)%len(input)][1]),
                      max(input[i][1], input[(i+1)%len(input)][1])]
            #print("Check line", rect_x, rect_y, line_x, line_y)
            if rect_x_line( rect_x, rect_y, line_x, line_y ):
                #print("Not admissible")
                admissible = False
                break

    # TODO: Check rect midpoint in loop

    return admissible
```
