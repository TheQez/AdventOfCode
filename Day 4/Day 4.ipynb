```python
f = open("input.txt")

input = []
for line in f:
    input.append([str(x) for x in line.rstrip()])

f.close()

print(input)
```

```python
def num_adj( grid, x, y ):
    total = 0
    for i in range(max(x-1, 0), min(x+2, len(grid))):
        for j in range(max(y-1, 0), min(y+2, len(grid[0]))):
            if (i != x or j != y) and grid[i][j] == '@':
                total += 1
    return total
```
