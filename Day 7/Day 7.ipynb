```python
input = []

f = open("input.txt")

for line in f:
    input.append([char for char in line.rstrip()])

f.close()

print(input)
```

```python
starting_pos = 0
split_loc = []

for i in range(0, len(input[0])):
    if input[0][i] == 'S':
        starting_pos = i
        break

print(starting_pos)

for i in range(1, len(input)):
    splitters = []
    for j in range(0, len(input[i])):
        if input[i][j] == '^':
            splitters.append(j)
    split_loc.append(splitters)

print(split_loc)
```

Part 1

```python
def propagate( prev_beams, splitters ):
    new_beams = set()
    num_splits = 0
    for beam in prev_beams:
        if beam in splitters:
            new_beams.add(beam-1)
            new_beams.add(beam+1)
            num_splits += 1
        else:
            new_beams.add(beam)

    return new_beams, num_splits

propagate( [1, 3], [1, 3] )
```

```python
curr_beams = set()
curr_beams.add(starting_pos)
tot_splits = 0

for splitters in split_loc:
    curr_beams, this_splits = propagate(curr_beams, splitters)
    tot_splits += this_splits

print(tot_splits)
```

Part 2

```python
from collections import defaultdict

def propagate_quantum( prev_routes, splitters ):
    new_routes = defaultdict(int)
    for pos, num_routes in prev_routes.items():
        if pos in splitters:
            new_routes[pos-1] += num_routes
            new_routes[pos+1] += num_routes
        else:
            new_routes[pos] += num_routes

    return new_routes

temp_dict = defaultdict(int)
temp_dict[2] = 1
temp_dict[4] = 2
propagate_quantum(temp_dict, [2, 4])
```

```python
curr_routes = defaultdict(int)
curr_routes[starting_pos] = 1

for splitters in split_loc:
    curr_routes = propagate_quantum(curr_routes, splitters)

tot_routes = 0
for _, num_routes in curr_routes.items():
    tot_routes += num_routes

print(tot_routes)
```
