```python
input = []

f = open("input.txt")

for line in f:
    input.append(line.rstrip().split())

f.close()

print(input)
```

Part 1

```python
total = 0
tot_arr = []

for i in range(0, len(input[0])):
    this_tot = 0
    if input[-1][i] == '+':
        this_tot = 0
        for j in range(0, len(input)-1):
            this_tot += int(input[j][i])
    else:
        this_tot = 1
        for j in range(0, len(input)-1):
            this_tot *= int(input[j][i])
    tot_arr.append(this_tot)

for num in tot_arr:
    total += num

print(tot_arr)
print(total)
```

Part 2

```python
grid_input = []
sizes = []

for i in range(0, len(input[0])):
    max_size = 0
    for j in range(0, len(input)):
        max_size = max(max_size, len(input[j][i]))
    sizes.append(max_size)

f = open("input.txt")

for line in f:
    idx = 0
    this_row = []
    for i in range(0, len(sizes)):
        this_row.append(line[idx:idx+sizes[i]].strip('\n').ljust(sizes[i], ' '))
        idx += sizes[i]+1
    grid_input.append(this_row)

f.close()

print(sizes)
print(grid_input)
```

```python
total = 0
tot_arr = []

for col in range(0, len(grid_input[0])):
    this_tot = 0
    if grid_input[-1][col].strip() == '+':
        this_tot = 0
        for num_idx in range(0, sizes[col]):
            str_num = ''
            for row in range(0, len(grid_input)-1):
                str_num += grid_input[row][col][num_idx]
            this_tot += int(str_num)
    else:
        this_tot = 1
        for num_idx in range(0, sizes[col]):
            str_num = ''
            for row in range(0, len(grid_input)-1):
                str_num += grid_input[row][col][num_idx]
            this_tot *= int(str_num)
    tot_arr.append(this_tot)

for num in tot_arr:
    total += num

print(total)
```
