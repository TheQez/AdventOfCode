```python
safe_range_low = []
safe_range_high = []
input = []

f = open("input.txt")

for line in f:
    if '-' in line:
        low, high = line.rstrip().split('-')
        safe_range_low.append(int(low))
        safe_range_high.append(int(high))
    elif line != '\n':
        input.append(int(line.rstrip()))

f.close()
```

Part 1

```python
total = 0

for i in input:
    for j in range(0, len(safe_range_low)):
        if i >= safe_range_low[j] and i <= safe_range_high[j]:
            total += 1
            break

print(total)
```

Part 2

```python
# Return the union of two intervals if their union is not disjoint
def merge_itvl( itvl0, itvl1 ):
    if ( ( itvl1[1] >= itvl0[0] and itvl1[1] <= itvl0[1] ) or
         ( itvl0[1] >= itvl1[0] and itvl0[1] <= itvl1[1] ) ):
        return [min(itvl0[0], itvl1[0]), max(itvl0[1], itvl1[1])]
    else:
        return []

merge_itvl([0, 7], [2, 3])
```

```python
disjoint_input = []
for i in range(0, len(safe_range_low)):
    disjoint_input.append([safe_range_low[i], safe_range_high[i]])

for i in range(0, len(disjoint_input)):
    for j in range(i+1, len(disjoint_input)):
        merged = merge_itvl(disjoint_input[i], disjoint_input[j])
        if len(merged) != 0:
            disjoint_input[i] = []
            disjoint_input[j] = merged
            break
```

```python
total = 0

for i in range(0, len(disjoint_input)):
    if len(disjoint_input[i]) != 0:
        total += disjoint_input[i][1] - disjoint_input[i][0] + 1

print(total)
```
