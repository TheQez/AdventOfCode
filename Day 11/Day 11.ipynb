```python
DAG = {}

f = open("input.txt")

for line in f:
    split = line.split()
    DAG[split[0][:-1]] = split[1:]

f.close()

DAG['out'] = []

print(DAG)
```

```python
inv_DAG = {}

for node in DAG:
    inv_DAG[node] = []

for from_node in DAG:
    for to_node in DAG[from_node]:
        inv_DAG[to_node].append(from_node)

print(inv_DAG)
```

Part 1

```python
def n_paths( start_node, end_node, DAG, inv_DAG ):
    n_paths = {}

    for node in DAG:
        n_paths[node] = -1
    n_paths[start_node] = 1

    while True:
        for to_node in inv_DAG:
            if n_paths[to_node] == -1:
                can_fill = True
                for from_node in inv_DAG[to_node]:
                    if n_paths[from_node] == -1:
                        can_fill = False
                        break

                if ( can_fill ):
                    n_paths[to_node] = 0
                    for from_node in inv_DAG[to_node]:
                        n_paths[to_node] += n_paths[from_node]

        if n_paths[end_node] != -1:
            return n_paths[end_node]
```

```python
print( n_paths( 'you', 'out', DAG, inv_DAG ) )
```

Part 2

```python
route_1 = (n_paths( 'svr', 'dac', DAG, inv_DAG ) *
           n_paths( 'dac', 'fft', DAG, inv_DAG ) *
           n_paths( 'fft', 'out', DAG, inv_DAG ))

route_2 = (n_paths( 'svr', 'fft', DAG, inv_DAG ) *
           n_paths( 'fft', 'dac', DAG, inv_DAG ) *
           n_paths( 'dac', 'out', DAG, inv_DAG ))

print(route_1 + route_2)
```
