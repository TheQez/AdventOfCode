```python
f = open("input.txt")

interval_str = []
itvl_low = []
itvl_high = []
for line in f:
    interval_str = line.split(',')

for itvl in interval_str:
    print(itvl)
    arr = itvl.split('-')
    itvl_low.append(int(arr[0]))
    itvl_high.append(int(arr[1]))

f.close()
```

```python
def is_invalid( num ):
    str_num = str(num)
    if ( len(str_num) % 2 != 0 ):
        return False
    mid = len(str_num) // 2
    first_half = str_num[:mid]
    second_half = str_num[mid:]

    return first_half == second_half
```

```python
def is_invalid_2( num ):
    str_num = str(num)

    for i in range(2, len(str_num)+1):
        if len(str_num) % i == 0:
            matches = True
            gap = len(str_num) // i
            match_str = str_num[:gap]

            for j in range(1, len(str_num)//gap):
                if match_str != str_num[j*gap:(j+1)*gap]:
                    matches = False
            if matches:
                return True

    return False
```

```python
total = 0

for i in range(0, len(itvl_low)):
    for num in range(itvl_low[i], itvl_high[i]):
        if ( is_invalid_2(num) ):
            total += num

print(total)
```
